# Obtaining PDFs from the basic Celonis documentation (PQL, platform related, etc). 

```python
# WORKING!!


from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os

# Set up the directory to save screenshots
save_dir = '/Users/adri/Desktop/Pictures/PQL Doc'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Initialize the WebDriver
driver = webdriver.Chrome()

# Open the first page
url = 'https://docs.celonis.com/en/getting-started.html'
driver.get(url)

# Wait for the page to load
time.sleep(5)

# Loop until there are no more "Next" buttons
page_num = 1
while True:
    # Save the current page as a screenshot
    screenshot_path = os.path.join(save_dir, f"page_{page_num}.png")
    driver.save_screenshot(screenshot_path)
    print(f"Saved screenshot: {screenshot_path}")

    try:
        # Use a CSS selector to find the "Next" button
        next_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "li.next a"))
        )

        # Scroll the "Next" button into view
        driver.execute_script("arguments[0].scrollIntoView(true);", next_button)

        # Try to click the "Next" button if it's clickable
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "li.next a"))
        )

        # Click the "Next" button
        next_button.click()
        print(f"Clicked 'Next' button, moving to page {page_num + 1}")
        page_num += 1

        # Wait for the next page to load fully
        time.sleep(5)

    except Exception as e:
        # If the element is not clickable, navigate using the href attribute
        try:
            href = next_button.get_attribute("href")
            print(f"Navigating directly to: {href}")
            driver.get(href)
            page_num += 1
            time.sleep(5)
        except:
            print(f"No more 'Next' buttons found, ending loop. Exception: {str(e)}")
            break

# Close the browser when done
driver.quit()
```
